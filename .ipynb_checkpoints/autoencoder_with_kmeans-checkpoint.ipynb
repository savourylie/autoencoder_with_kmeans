```python
from collections import namedtuple
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F
from collections import defaultdict
import sys

torch.cuda.set_device(3)

class AutoencoderKmeans(nn.Module):
    def __init__(self, input_dim, encoding_dim=10, num_clusters=10):
        super().__init__()
        # Global timestamp
        self.t = 0
        
        # Hyper parameters
        self.num_clusters = num_clusters
        self.encoding_dim = encoding_dim
        
        # Define AE architecture
        self.fc1 = nn.Linear(input_dim, self.encoding_dim)
        self.fc2 = nn.Linear(self.encoding_dim, input_dim)
        
        self.enc1 = nn.Linear(input_dim, 1000)
        nn.init.xavier_uniform(self.enc1.weight)
        self.enc2 = nn.Linear(1000, 250)
        nn.init.xavier_uniform(self.enc2.weight)
        self.enc3 = nn.Linear(250, 50)
        nn.init.xavier_uniform(self.enc3.weight)
        self.enc4 = nn.Linear(50, self.encoding_dim)
        nn.init.xavier_uniform(self.enc4.weight)
        self.dec1 = nn.Linear(self.encoding_dim, 50)
        nn.init.xavier_uniform(self.dec1.weight)
        self.dec2 = nn.Linear(50, 250)
        nn.init.xavier_uniform(self.dec2.weight)
        self.dec3 = nn.Linear(250, 1000)
        nn.init.xavier_uniform(self.dec3.weight)
        self.dec4 = nn.Linear(1000, input_dim)
        nn.init.xavier_uniform(self.dec4.weight)

        # Initialize cluster centers
#         self.centers = torch.from_numpy(np.random.uniform(0, 1, (self.num_clusters, self.encoding_dim))).type(torch.FloatTensor)
        self.centers = None
        self.clusters = None
        self.AKoutput = namedtuple('AKoutput', 'k_means_loss clusters data_hidden output')
        
    def forward(self, x):
        if self.training:
#             x = self.fc1(x)
#             x = F.tanh(x)
#             data_hidden = x.clone()
#             x = self.fc2(x)

            x = self.enc1(x)
            
            x = F.elu(x)
            x = F.dropout(x, 0.5)
            x = self.enc2(x)
            x = F.elu(x)
            x = F.dropout(x, 0.5)
            x = self.enc3(x)
            x = F.elu(x)
            x = F.dropout(x, 0.5)
            x = self.enc4(x)
            x = F.dropout(x, 0.3)
            
            data_hidden = x.clone()
            
            if self.t == 0:
                self.init_clusters(data_hidden)
                k_means_loss = np.nan
                current_clusters = None
            
            if self.t > 0:
                k_means_loss = self.kmeans_opt(data_hidden)
                current_clusters = self.get_clusters(data_hidden)

            x = self.dec1(x)
            x = F.elu(x)
            x = F.dropout(x, 0.5)
            x = self.dec2(x)
            x = F.elu(x)
            x = F.dropout(x, 0.5)
            x = self.dec3(x)
            x = F.elu(x)
            x = F.dropout(x, 0.5)
            x = self.dec4(x)
            
            output = x

#             print("K-means loss {0} at t = {1}".format(np.float(k_means_loss), self.t))
                  
            output_tuple = self.AKoutput(k_means_loss=k_means_loss, clusters=current_clusters, data_hidden=data_hidden, output=output)
            
            self.t += 1
            
            return output_tuple
    
        else:
#             x = self.fc1(x)
#             x = F.sigmoid(x)
#             data_hidden = x.clone()
#             x = self.fc2(x)
#             x = F.sigmoid(x)

            x = self.enc1(x)
            x = F.elu(x)
            x = self.enc2(x)
            x = F.elu(x)
            x = self.enc3(x)
            x = F.elu(x)
            x = self.enc4(x)

            data_hidden = x.clone()
            k_means_loss = self.kmeans_opt(data_hidden)
#             k_means_loss = 0

            x = self.dec1(x)
            x = F.elu(x)
            x = self.dec2(x)
            x = F.elu(x)
            x = self.dec3(x)
            x = F.elu(x)
            x = self.dec4(x)
            
            output = x
            
            current_clusters = self.get_clusters(data_hidden)
            
            output_tuple = self.AKoutput(k_means_loss=k_means_loss, clusters=current_clusters, data_hidden=data_hidden, output=output)
            
            return output_tuple
        
    def kmeans_opt(self, dataset):
        if self.training:
            self.centers = self.update_centers(dataset, self.clusters)
            self.clusters = self.get_clusters(dataset)
            k_means_loss = self.get_kmeans_loss(dataset, self.clusters)
            
        else:
            self.clusters = self.get_clusters(dataset)
            k_means_loss = self.get_kmeans_loss(dataset, self.clusters)
            
        return k_means_loss
            
    
    def init_clusters(self, dataset):
        self.clusters = np.random.randint(low=0, high=10, size=dataset.shape[0]).tolist()
        
    def get_clusters(self, dataset):
        clusters = []
        
        if torch.cuda.is_available():
            for data in dataset:
                clusters.append(((data.data - self.centers.cuda())**2).sum(1).min(0)[1][0])
        else:
            for data in dataset:
                clusters.append(((data.data - self.centers)**2).sum(1).min(0)[1][0])
        
        return clusters
    
    def get_kmeans_loss(self, dataset, clusters):
        loss = [0 for i in range(self.num_clusters)]
        
        if torch.cuda.is_available():
            for i, cluster in enumerate(clusters):
                loss[cluster] += ((dataset[i] - Variable(self.centers[cluster], requires_grad=False).cuda())**2).sum()

        else:
            for i, cluster in enumerate(clusters):
                loss[cluster] += ((dataset[i] - Variable(self.centers[cluster], requires_grad=False))**2).sum()
            
        return sum(loss) / len(loss)
    
    def update_centers(self, dataset, clusters):
        data_by_cluster = [[] for i in range(self.num_clusters)]
        
        for i, cluster in enumerate(clusters):
            data_by_cluster[cluster].append(dataset[i])
        
        if torch.cuda.is_available():
            centers = torch.from_numpy(np.zeros((self.num_clusters, self.encoding_dim))).type(torch.FloatTensor).cuda()
        else:
            centers = torch.from_numpy(np.zeros((self.num_clusters, self.encoding_dim))).type(torch.FloatTensor)
        
        for i, data in enumerate(data_by_cluster):
            centers[i] = (sum(data) / len(data)).data
        
        return centers
        
    @staticmethod
    def __square_loss(v1, v2):
        loss = ((v1 - v2)**2).sum(1)[0]
        
        return loss        
```

```python
input_dim = 784

model = AutoencoderKmeans(input_dim, encoding_dim=10)

if torch.cuda.is_available():
    model.cuda()
```

```python
num_layers = len(list(model.parameters()))

for i in range(num_layers):
    print(list(model.parameters())[i].size())
```

```python
criterion = nn.MSELoss()
learning_rate = 0.00001
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
lambd = 0.1
```

```python
num_epochs = 3000
```

```python
recon_losses = []
kmeans_losses = []
recon_losses_test = []
kmeans_losses_test = []
# kmeans_lambda = 0.001
kmeans_lambda = 10
batch_size = 1000

# GD
for epoch in range(num_epochs):
    if torch.cuda.is_available():
        x_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor).cuda()
        y_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor).cuda()
        x = Variable(x_torch).cuda()
        y = Variable(y_torch).cuda()
    else:
        x_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor)
        y_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor)
        x = Variable(x_torch)
        y = Variable(y_torch)

    optimizer.zero_grad()

    model.train()

    output_tuple = model(x)

    cost = criterion(output_tuple.output, y) + kmeans_lambda * output_tuple.k_means_loss

#         cost = criterion(output_tuple.output, y)

    cost.backward()

    optimizer.step()

    # Training loss
    if torch.cuda.is_available():
        x_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor).cuda()
        y_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor).cuda()
        x = Variable(x_torch).cuda()
        y = Variable(x_torch).cuda()
    else:
        x_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor)
        y_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor)
        x = Variable(x_torch)
        y = Variable(x_torch)

    model.train()
    output_tuple = model(x)
        
    recon_loss = float(((y_torch - output_tuple.output.data)**2).mean())
    kmeans_loss = float(kmeans_lambda * output_tuple.k_means_loss)
    recon_losses.append(recon_loss)
    kmeans_losses.append(kmeans_loss)

    if torch.cuda.is_available():
            x_torch_test = torch.from_numpy(X_test.values).type(torch.FloatTensor).cuda()
            y_torch_test = torch.from_numpy(X_test.values).type(torch.FloatTensor).cuda()
            x_test = Variable(x_torch_test).cuda()
            y_test = Variable(x_torch_test).cuda()
    else:
        x_torch_test = torch.from_numpy(X_test.values).type(torch.FloatTensor)
        y_torch_test = torch.from_numpy(X_test.values).type(torch.FloatTensor)
        x_test = Variable(x_torch_test)
        y_test = Variable(x_torch_test)

    model.eval()
    output_tuple_test = model(x_test)
    recon_loss_test = float(((y_torch_test - output_tuple_test.output.data)**2).mean())
    kmeans_loss_test = float(kmeans_lambda * output_tuple_test.k_means_loss)
    recon_losses_test.append(recon_loss_test)
    kmeans_losses_test.append(kmeans_loss_test)
    
    if epoch % 10 == 0:
        print("Epoch {}".format(epoch))
        print("Reconstruction Loss: {0}, K-means Loss: {1}".format(recon_loss, kmeans_loss))
        print("Validation: R-Loss: {0}, K-means Loss: {1}".format(recon_loss_test, kmeans_loss_test))

# Mini batch
# for epoch in range(num_epochs):
#     for i in range(0, len(X_train), batch_size):
#         X_mini, y_mini = X_train.values[i:i+batch_size], y_train.values[i:i+batch_size]

#         if torch.cuda.is_available():
#             x_torch = torch.from_numpy(X_mini).type(torch.FloatTensor).cuda()
#             y_torch = torch.from_numpy(X_mini).type(torch.FloatTensor).cuda()
#             x = Variable(x_torch).cuda()
#             y = Variable(y_torch).cuda()
#         else:
#             x_torch = torch.from_numpy(X_mini).type(torch.FloatTensor)
#             y_torch = torch.from_numpy(X_mini).type(torch.FloatTensor)
#             x = Variable(x_torch)
#             y = Variable(y_torch)

#         optimizer.zero_grad()

#         model.train()
        
#         output_tuple = model(x)

#         cost = criterion(output_tuple.output, y) + kmeans_lambda * output_tuple.k_means_loss
    
# #         cost = criterion(output_tuple.output, y)
        
#         cost.backward()

#         optimizer.step()

#     # Training loss
#     if torch.cuda.is_available():
#         x_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor).cuda()
#         y_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor).cuda()
#         x = Variable(x_torch).cuda()
#         y = Variable(x_torch).cuda()
#     else:
#         x_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor)
#         y_torch = torch.from_numpy(X_train.values).type(torch.FloatTensor)
#         x = Variable(x_torch)
#         y = Variable(x_torch)

#     model.train()
#     output_tuple = model(x)
        
#     recon_loss = float(((y_torch - output_tuple.output.data)**2).mean())
#     kmeans_loss = float(kmeans_lambda * output_tuple.k_means_loss)
#     recon_losses.append(recon_loss)
#     kmeans_losses.append(kmeans_loss)

#     if torch.cuda.is_available():
#             x_torch_val = torch.from_numpy(X_val.values).type(torch.FloatTensor).cuda()
#             y_torch_val = torch.from_numpy(X_val.values).type(torch.FloatTensor).cuda()
#             x_val = Variable(x_torch_val).cuda()
#             y_val = Variable(x_torch_val).cuda()
#     else:
#         x_torch_val = torch.from_numpy(X_val.values).type(torch.FloatTensor)
#         y_torch_val = torch.from_numpy(X_val.values).type(torch.FloatTensor)
#         x_val = Variable(x_torch_val)
#         y_val = Variable(x_torch_val)

#     model.eval()
#     output_tuple_val = model(x_val)
#     recon_loss_val = float(((y_torch_val - output_tuple_val.output.data)**2).mean())
#     kmeans_loss_val = float(kmeans_lambda * output_tuple_val.k_means_loss)
#     recon_losses_val.append(recon_loss_val)
#     kmeans_losses_val.append(kmeans_loss_val)
    
#     if epoch % 10 == 0:
#         print("Epoch {}".format(epoch))
#         print("Reconstruction Loss: {0}, K-means Loss: {1}".format(recon_loss, kmeans_loss))
#         print("Validation: R-Loss: {0}, K-means Loss: {1}".format(recon_loss_val, kmeans_loss_val))
```
